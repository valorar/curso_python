## Resuelto 14: Descuento especial
* Este es un caso de "código roto"
* Una tienda aplica un descuento del 10 % solo a clientes VIP. Pero el programador novato lo hizo mal.
* **Código roto:**
  ```python
  def precio_final(precio, es_vip):
      return precio * 0.90
      if es_vip:
          return precio * 0.90
      return precio
  ```
* Ejemplos:
    1. precio_final(100, False) ➞ 100
    2. precio_final(100, True) ➞ 90.0
    3. precio_final(50, True) ➞ 45.0


````python
# Método 1: corrigiendo con un condicional if/else
def precio_final(precio, es_vip):
    if es_vip:
        return precio * 0.90
    return precio
````

````python
# Método 2: con un ternario
def precio_final(precio, es_vip):
    return precio * 0.90 if es_vip else precio
````

````python
print(precio_final(100, False))
print(precio_final(100, True))
print(precio_final(50, True))
````

## Resuelto 20: Puntuación más alta del torneo

* En un torneo de videojuegos se registran las puntuaciones de todas las partidas jugadas.
* Cree una función que reciba una lista de puntuaciones y devuelva la más alta.
* Ejemplos:
    1. puntuacion_alta([120, 350, 90, 480]) ➞ 480
    2. puntuacion_alta([1000, 999, 1000]) ➞ 1000
    3. puntuacion_alta([50]) ➞ 50

````python
# Método 1: usando la función max()
def puntuacion_alta(lista):
    return max(lista)
````

````python
# Método 2: recorriendo la lista manualmente
def puntuacion_alta(lista):
    maxima = lista[0]
    for puntuacion in lista:
        if puntuacion > maxima:
            maxima = puntuacion
    return maxima
````

## Resuelto 21: Temperatura más baja de la semana

* Una estación meteorológica registra la temperatura mínima diaria durante una semana.
* Los valores pueden contener decimales y ser negativos.
* Cree una función que reciba la lista de temperaturas y devuelva la más baja.
* Ejemplos:
    1. temperatura_minima([12, 5, -3, 8, 15, 0, -1]) ➞ -3
    2. temperatura_minima([4.5, -2.3, 0, 1.1]) ➞ -2.3
    3. temperatura_minima([7, 7, 7]) ➞ 7

````python
# Método 1: usando la función min()
def temperatura_minima(lista):
    return min(lista)
````

````python
# Método 2: recorriendo la lista manualmente
def temperatura_minima(lista):
    minima = lista[0]
    for temp in lista:
        if temp < minima:
            minima = temp
    return minima
````

* En muchas competiciones deportivas (patinaje, gimnasia, saltos), los jueces descartan la puntuación más alta y la más baja para evitar que una nota extrema favorezca o perjudique injustamente al deportista. La puntuación final es la media del resto de notas.

* Si hay menos de 3 puntuaciones, devuelve `None` (no se puede aplicar el sistema).

* Ejemplos:
    1. media_central([10, 20, 30, 40, 50, 50]) ➞ 35.0
    2. media_central([5, 5, 5, 5, 3]) ➞ 5.0
    3. media_central([1, 100]) ➞ None
    4. media_central([]) ➞ None

````python
# Método 1: eliminando el mínimo y el máximo con remove(), luego calculando la media
def media_central(lista):
    if len(lista) < 3:
        return None
    copia = lista.copy()
    copia.remove(min(copia))
    copia.remove(max(copia))
    return sum(copia) / len(copia)
````

````python
# Método 2: ordenando la lista y descartando los extremos
def media_central(lista):
    if len(lista) < 3:
        return None
    ordenada = sorted(lista)
    centro = ordenada[1:-1]
    return sum(centro) / len(centro)
````

* Una central de alarmas recibe las señales de un número variable de sensores. Recuerde que un número indeterminado de parámetros se recoge con la sintaxis **\*args**.

* Los argumentos pueden ser cualquier tipo de dato en Python: se evalúan como booleanos.

* Ejemplos:
    1. hay_alerta(False, False, True) ➞ True
    2. hay_alerta(0, "", []) ➞ False
    3. hay_alerta("humo") ➞ True
    4. hay_alerta(None, 0, False) ➞ False

### Comentario

En Python, todos los objetos tienen un valor booleano intrínseco cuando se evalúan en un contexto lógico (como un `if` o al usar funciones lógicas). Esto se conoce como valores **Truthy** (verdaderos) y **Falsy** (falsos):

* **Valores Falsy (evalúan como `False`):** `False`, `None`, el cero (`0`, `0.0`), y cualquier colección o cadena vacía (`""`, `[]`, `{}`, `()`).
* **Valores Truthy (evalúan como `True`):** Cualquier cosa que no sea Falsy. Esto incluye números distintos de cero y **cadenas de texto que no estén vacías**.

Como `"humo"` es una cadena de texto con contenido (no está vacía), Python la evalúa como **`True`**.


````python
# Método 1: usando any()
def hay_alerta(*args):
    return any(args)
````

````python
# Método 2: recorriendo los argumentos manualmente
def hay_alerta(*args):
    for arg in args:
        if arg:
            return True
    return False
````

````python
print(hay_alerta(False, False, True))
print(hay_alerta(0, "", []))
print(hay_alerta("humo"))
print(hay_alerta(None, 0, False))
````
